```python
# Importar las bibliotecas 
import pandas as pd  
import numpy as np   
import seaborn as sns  # Creacin grficos estadsticos
import matplotlib.pyplot as plt  # Creacin de grficos en dos dimensiones

# Ruta del dataset
ruta_dataset = r"C:/DATE_SET_CENTRO_COMERCIAL/Mall_Customers.csv"

# Cargar el dataset
df = pd.read_csv(ruta_dataset)

# Eliminar la columna 'CustomerID'
df = df.drop(columns=['CustomerID'])

# Convertir la columna 'Gender' a tipo string
df['Gender'] = df['Gender'].astype(str)

# Codificar la columna 'Gender' a valores numricos
df['Gender'] = df['Gender'].map({'Male': 0, 'Female': 1})

# Mostrar las primeras filas del dataset para verificar la conversin
print(df.head())

# Visualizacin de la distribucin de datos
plt.figure(figsize=(12, 6))
sns.histplot(df['Annual Income (k$)'], bins=30, kde=True)
plt.title('Distribucin de Ingresos Anuales')
plt.xlabel('Ingresos Anuales (k$)')
plt.ylabel('Frecuencia')
plt.show()

plt.figure(figsize=(12, 6))
sns.histplot(df['Spending Score (1-100)'], bins=30, kde=True)
plt.title('Distribucin del Puntaje de Gastos')
plt.xlabel('Puntaje de Gastos (1-100)')
plt.ylabel('Frecuencia')
plt.show()

# Grfico de dispersin para identificar relaciones
plt.figure(figsize=(10, 6))
sns.scatterplot(data=df, x='Annual Income (k$)', y='Spending Score (1-100)')
plt.title('Relacin entre Ingresos Anuales y Puntaje de Gastos')
plt.xlabel('Ingresos Anuales (k$)')
plt.ylabel('Puntaje de Gastos (1-100)')
plt.show()

# Identificacin de valores atpicos con boxplots
plt.figure(figsize=(12, 6))
sns.boxplot(data=df[['Annual Income (k$)', 'Spending Score (1-100)']])
plt.title('Boxplot de Ingresos Anuales y Puntaje de Gastos')
plt.ylabel('Valor')
plt.show()

# Matriz de correlacin
plt.figure(figsize=(8, 6))
correlation_matrix = df.corr()
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', fmt='.2f')
plt.title('Matriz de Correlacin')
plt.show()


# Analisis de la grafica de Distribucin de Ingresos Anuales y Puntaje de Gasto:
# Distribucin Normal: Ambos muestran una distribucin aproximadamente normal con concentraciones en ciertos rangos.
# Clientes de Altos Ingresos: Hay un pico en ingresos altos, indicando un segmento con alto poder adquisitivo.
# Diversidad en el Puntaje de Gasto: La distribucin del puntaje de gasto es ms uniforme, reflejando variedad en hbitos de consumo.


#Analisis de la Relacin entre Ingresos Anuales y Puntaje de Gasto:
#Correlacin Positiva: Se observa una relacin positiva moderada,
#donde mayores ingresos tienden a asociarse con un puntaje de gasto ms alto, aunque no es perfecta.


#Analisis de Segmentacin por Gnero:
#Similitudes: No se encuentran diferencias significativas en ingresos y gasto entre hombres y mujeres.

# Analisis del boxplot por gnero no muestra diferencias significativas en la distribucin del ingreso anual y el puntaje de gasto entre hombres y mujeres.
# Esto sugiere que ambos gneros tienen un comportamiento de compra similar en trminos de estos dos factores.

# Analisis de la Matriz de correlacion
# Correlacin positiva: El grfico de correlacin muestra una relacin positiva moderada entre el ingreso anual y el puntaje de gasto.
# Esto significa que, en general, los clientes con mayores ingresos tienden a tener un puntaje de gasto ms alto. Sin embargo,
# esta relacin no es perfecta, lo que indica que otros factores tambin influyen en el comportamiento de compra.

```
