```python
# Importar las bibliotecas 
import pandas as pd  
import numpy as np   
import seaborn as sns  # Creación gráficos estadísticos
import matplotlib.pyplot as plt  # Creación de gráficos en dos dimensiones

# Ruta del dataset
ruta_dataset = r"C:/DATE_SET_CENTRO_COMERCIAL/Mall_Customers.csv"

# Cargar el dataset
df = pd.read_csv(ruta_dataset)

# Eliminar la columna 'CustomerID'
df = df.drop(columns=['CustomerID'])

# Convertir la columna 'Gender' a tipo string
df['Gender'] = df['Gender'].astype(str)

# Codificar la columna 'Gender' a valores numéricos
df['Gender'] = df['Gender'].map({'Male': 0, 'Female': 1})

# Mostrar las primeras filas del dataset para verificar la conversión
print(df.head())

# Visualización de la distribución de datos
plt.figure(figsize=(12, 6))
sns.histplot(df['Annual Income (k$)'], bins=30, kde=True)
plt.title('Distribución de Ingresos Anuales')
plt.xlabel('Ingresos Anuales (k$)')
plt.ylabel('Frecuencia')
plt.show()

plt.figure(figsize=(12, 6))
sns.histplot(df['Spending Score (1-100)'], bins=30, kde=True)
plt.title('Distribución del Puntaje de Gastos')
plt.xlabel('Puntaje de Gastos (1-100)')
plt.ylabel('Frecuencia')
plt.show()

# Gráfico de dispersión para identificar relaciones
plt.figure(figsize=(10, 6))
sns.scatterplot(data=df, x='Annual Income (k$)', y='Spending Score (1-100)')
plt.title('Relación entre Ingresos Anuales y Puntaje de Gastos')
plt.xlabel('Ingresos Anuales (k$)')
plt.ylabel('Puntaje de Gastos (1-100)')
plt.show()

# Identificación de valores atípicos con boxplots
plt.figure(figsize=(12, 6))
sns.boxplot(data=df[['Annual Income (k$)', 'Spending Score (1-100)']])
plt.title('Boxplot de Ingresos Anuales y Puntaje de Gastos')
plt.ylabel('Valor')
plt.show()

# Matriz de correlación
plt.figure(figsize=(8, 6))
correlation_matrix = df.corr()
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', fmt='.2f')
plt.title('Matriz de Correlación')
plt.show()


# Analisis de la grafica de Distribución de Ingresos Anuales y Puntaje de Gasto:
# Distribución Normal: Ambos muestran una distribución aproximadamente normal con concentraciones en ciertos rangos.
# Clientes de Altos Ingresos: Hay un pico en ingresos altos, indicando un segmento con alto poder adquisitivo.
# Diversidad en el Puntaje de Gasto: La distribución del puntaje de gasto es más uniforme, reflejando variedad en hábitos de consumo.


#Analisis de la Relación entre Ingresos Anuales y Puntaje de Gasto:
#Correlación Positiva: Se observa una relación positiva moderada,
#donde mayores ingresos tienden a asociarse con un puntaje de gasto más alto, aunque no es perfecta.


#Analisis de Segmentación por Género:
#Similitudes: No se encuentran diferencias significativas en ingresos y gasto entre hombres y mujeres.

# Analisis del boxplot por género no muestra diferencias significativas en la distribución del ingreso anual y el puntaje de gasto entre hombres y mujeres.
# Esto sugiere que ambos géneros tienen un comportamiento de compra similar en términos de estos dos factores.

# Analisis de la Matriz de correlacion
# Correlación positiva: El gráfico de correlación muestra una relación positiva moderada entre el ingreso anual y el puntaje de gasto.
# Esto significa que, en general, los clientes con mayores ingresos tienden a tener un puntaje de gasto más alto. Sin embargo,
# esta relación no es perfecta, lo que indica que otros factores también influyen en el comportamiento de compra.

```
